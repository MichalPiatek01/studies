## Zadanie 1 (25 pkt) Listy
Napisz 3 funkcje (po 7 pkt każda). Funkcje mają 1 paramert formalny - sekwencję liczb naturalnych `N` (lista, krotka...). Na bazie liczb z `N` skonstuuj listę wartości 
wyrażeń $$\frac{k}{k + 7}$$ k-tych
elementów `N` i zwróć ją z funkcji.

* funkcja 1: wykorzystaj pętlę `while` i 
metodę `append`
* funkcja 2: wykorzystaj pętlę `for` i
konkatenację
* funkcja 3: wykorzystaj wyrażenie listowe

Dodatkowo
* docstring (2 pkt)
* testy każdej z funkcji (2 pkt)

```python
def zadanie_1_while(*args):
    '''Funkcja zadanie_1_while 
    
    Funkcja wykorzystuje pętle while i
    metodę append do skonstruowania
    listy wartości wyrażenia k/(k+7)
    
    Args:
        args: list[int] - lista argumentów 
    
    Returns:
        list[float]
    '''
    L = []
    i = 0
    while i < len(args):
        L.append(args[i]/(args[i]+7))
        i +=1
    return L

print(zadanie_1_while(1,2,3),zadanie_1_while(5,6,7),sep='\n')
```

```python
def zadanie_1_for(*args):
    '''Funkcja zadanie_1_for 
    
    Funkcja wykorzystuje pętle for i
    metodę konkatenacji do skonstruowania
    listy wartości wyrażenia k/(k+7)
    
    Args:
        args: list[int] - lista argumentów 
    
    Returns:
        list[float]
    '''
    L = []
    for i in args:
        L += [i/(i+7)]
    return L

print(zadanie_1_for(1,2,3),zadanie_1_for(5,6,7),sep='\n')
```

```python
def zadanie_1_wyrazenie_listowe(*args):
    '''Funkcja zadanie_1_for 
    
    Funkcja wykorzystuje wyrażenie listowe do skonstruowania
    listy wartości wyrażenia k/(k+7)
    
    Args:
        args: list[int] - lista argumentów 
    
    Returns:
        list[float] 
    '''
    return [i/(i+7) for i in args]

print(zadanie_1_wyrazenie_listowe(1,2,3),zadanie_1_wyrazenie_listowe(5,6,7),sep='\n')
```

## Zadanie 2 (25 pkt) TLD
Napisz funkcję `tld(url)` zwracającą nazwę Domeny Najwyższego Poziomu (TLD) z podanego [url](https://pl.wikipedia.org/wiki/Uniform_Resource_Locator) (10 pkt).

Istnieją 2 typy domen najwyższego poziomu:
* krajowe (ang. ccTLD – country code TLD), zawsze dwuliterowe; np. Polska: .pl
* funkcjonalne (ang. gTLD – generic TLD) np. .org, .biz, .info

Przykłady

    'https://us.edu.pl/instytut/ifiz' -> pl
    'google.com' -> com
    'ftp://zoom.zoo.cz/e/1/2/' -> cz
    'https://web.mit.edu/' -> edu


* Funkcja powinna informować, czy domena jest typu ccTLD czy gTLD (5 pkt).
* Zdekomponuj problem: 2 funkcje to 3 pkt, 3 funkcje to 5 pkt
* Docstring (2 pkt)
* Wykonaj trzy testy dla kilku różnych adresów (3 pkt) innych od podanych w przykładach powyżej.

```python
def usun_protokol(url):
    '''Funkcja usun_protokol
    
    Funkcja wykorzystuje wbudowaną funkcje find 
    do znalezienia protokołu po czym go usuwa
    
    Args:
        url : str - prawidłowy url
        
    Returns:
        str - url bez protokołu
    '''
    index = url.find(':')
    if index != -1:
        url = url[index+3:]
    return url

def wydziel_do_slasha(url):
    '''Funkcja wydziel_do_slasha
    
    Funkcja wykorzystuje wbudowaną funkcje find 
    do znalezienia '/' po czym usuwa od niego 
    aż do końca
    
    Args:
        url : str - prawidłowy url
        
    Returns:
        str - url bez zasobów strony
    '''
    index = url.find('/')
    if index != -1:
        url = url[:index]
    return url

def wydziel_od_kropki(url):
    '''Funkcja wydziel_od_kropki
    
    Funkcja wykorzystuje wbudowaną funkcje rindex
    do znalezienia ostatniej '.' po czym usuwa od
    niej aż do początku
    
    Args:
        url : str - prawidłowy url
        
    Returns:
        str - Domeny Najwyższego Poziomu
    
    '''
    index = url.rindex('.')
    if index != -1:
        url = url[index+1:]
    return url

def tld(url):
    '''Funckja tld
    
    Funkcja wykorzystuje funkcje pomocniczę  
    do znalezienia Domeny Najwyższego poziomu
    
    Args:
        url : str - prawidłowy url
        
    Returns:
        str - Domeny Najwyższego Poziomu
    '''
    url = usun_protokol(url)
    url = wydziel_do_slasha(url)
    url = wydziel_od_kropki(url)
    return url

print(tld('https://alfa.smcebi.us.edu.pl/user/michal.piatek/notebooks/zadania%20wdp/is21_wdp_kolokwium2_G7_G7_Pi%C4%85tek_Micha%C5%82.ipynb'),
      tld('https://twitter.com/home'),
      tld('https://usosweb.us.edu.pl/kontroler.php?_action=home/index'),sep='\n')
```

## Zadanie 3 (25 pkt) Maksymalna przerwa binarna

Przerwa binarna to sekwencja bitów o wartości `0` zawarta między dwoma wartościami `1` w bitowej reprezentacji liczb. Przykładowo liczba 

$20_{dec} = 10100_{bin}$

ma przerwę binarną o długości `1` zlokalizowaną pomiędzy pierwszym i trzecim jej bitem. Znajdziemy tam też dwa zera na końcu liczby na 4 i 5 bicie, ale nie tworzą one przerwy binarnej, ponieważ nie istnieje bit `1` po tej sekwencji zer. W tym przypadku możemy powiedzieć, że przerwa binarna liczby `20` wynosi `1`. Długość maksymalnej przerwy binarnej dla liczby 

$529_{dec} = 1000010001_{bin}$

wynosi jak widać `4`.

* Napisz funkcję obliczającą maksymalną przerwę binarną dla zadanej liczby (15 pkt). 
* Zdekomponuj problem: 2 funkcje to (3 pkt), więcej niż 2 funkcje to (5 pkt).
* Napisz minimum 3 testy dla tego zadania (3 pkt); punkty tylko za testy inne niż te powyżej.
* Napisz docstring (2 pkt).


```python
def dec_To_Bin(liczba):
    '''Funkcja dec_To_Bin
    
    Funkcja zwraca reprezentacje binarną liczby
    
    Args:
        liczba: int
        
    Returns:
        str - liczba w postaci binarnej
    '''
    return bin(liczba)[2:]
 
def znajdz_przerwe(liczbaB):
    '''Funkcja znajdz_przerwe
    
    Funkcja wykorzystuje petle for do znalezienia maksymalnej przewy binarnej
    
    Args:
        liczbaB: str - liczba w postaci binarnej
        
    Returns:
        int - Maksymalna przerwa binarna
    '''
    jedynki = 0
    zera = 0
    maxPrzerwa = 0
    for i in liczbaB:
        if i == '1':
            jedynki += 1
        elif i == '0':
            zera += 1
        if jedynki == 2:
            if zera > maxPrzerwa:
                maxPrzerwa = zera
            zera = 0
            jedynki = 1
    return maxPrzerwa

def przerwa_Binarna(liczba):
    '''Funkcja przerwa_Binarna
    
    Funkcja wykorzystuje funkcje pomocniczę
    do znalezienia maksymalnej przerwy binarnej
    
    Args:
        liczba: int
    
    Returns:
        int - Maksymalna przerwa binarna'''
    liczbaB = dec_To_Bin(liczba)
    print(liczbaB)
    return znajdz_przerwe(liczbaB)
 

print(przerwa_Binarna(300),przerwa_Binarna(756),przerwa_Binarna(1547))
```

## Zadanie 4 (25 pkt) compress
Napisz funkcję zwracającą słownik (5 pkt) będącą reprezentacją skompresowanego ciągu znaków (15 pkt).

Kompresja ma polegać na znalezieniu wszystkich spacji oraz przecinków, zapamiętania ich położenia w oryginalnym ciągu znaków, usunięcia ich z tego ciągu i w końcu stworzenia słownika zawierającego

    {'tekst': oryginalny_tekst_bez_spacji_i_przecinkow, 
    'spacje': krotka_z_miejscami_spacji, 
    'przecinki': krotka_z_miejscami_przecinkow}

Np:

    >>> tekst = 'Ala ma kota, ale nie ma psa'
    >>> compress(tekst)
    {'przecinki': (11,), 'tekst': 'Alamakotaaleniemapsa', 'spacje': (3, 6, 12, 16, 20, 23)}

Dodatkowo
* Zdekomponuj problem (5 pkt).
* Przeprowadź testy (3 pkt). 
* Nie zapomnij o docstringu (2 pkt).

Wskazówki
* Nie przejmuj się, jeżeli tekst ma inne znaki oprócz wymienionych - zajmij się tylko przecinkami, spacjami i "resztą"

```python
def zlicz_przecinki_i_spacje(tekst):
    '''Funkcja zlicz_przecinki_i_spacje
    
    Funkcja wykorzystuje pętle for do znalezienia
    indexów przecinków i spacji
    
    Args:
        tekst: str
        
    Returns:
    
        tuple(int)- indexy przecinków, tuple(int)- indexy spacji '''
    przecinki = []
    spacje = []
    for i in range(len(tekst)):
        if tekst[i] == ',':
            przecinki += [i]
        if tekst[i] == ' ':
            spacje += [i]
    przecinki = tuple(przecinki)
    spacje = tuple(spacje)
    return przecinki, spacje

def usun_przecinki_i_spacje(tekst):
    '''Funkcja usun_przecinki_i_spacje
    
    Funkcja wykorzystuje funkcję wbudowaną 
    replace do usunięcia przecinków i spacji
    
    Args:
        tekst: str
        
    Returns:
        dictionary
    '''
    tekst = tekst.replace(',','')
    tekst = tekst.replace(' ', '')
    return tekst

def slownik(tekst):
    przecinki, spacje = zlicz_przecinki_i_spacje(tekst)
    tekst = usun_przecinki_i_spacje(tekst)
    return {'przecinki':przecinki, 'tekst':tekst, 'spacje':spacje}

print(slownik('ala ma kota, a kot ma ale'),
      slownik('W porannej rosie Brudny,lecz świeży, Ubłocony melon. '),
      slownik('Pada wiosenny deszcz; Z dachu przecieka, Prosto do gniazda os.'),sep='\n')
```
