Proszę skonsturować hierarchię klas opisujących trójkąty

    Figura

    ......TrojkatProstakatny
    ......TrojkatRównoramienny
    .........TrojkatRównoboczny

    
Funkcjonalność

* obwód
* wysokość
* `czy_wiekszy_od` bazując na obwodzie

Na koniec proszę przeprowadzić testy klas: 
* 3 konkretne, różne instancje dla każdej klasy potomnej, prosze przejrzeć (uruchomić, przetestować) funkcje które państwo programują
* porównać każdy trójkąt z każym (z tych instancji)
* stworzyć kwadrat i prostokąt (kod na labie) i porównać (`czy_wiekszy_od`) ze stworzonymi instancjami

```python
import math

class Figura:
    def pole(self):
        pass
    
    def obwod(self):
        pass
    
    def wysokosc(self):
        pass
    
    def czy_wiekszy_od(self, other):
        return self.obwod() > other.obwod()


class Trojkat(Figura):
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
        
    def pole(self):
        p = (self.a + self.b + self.c) / 2
        return math.sqrt(p * (p - self.a) * (p - self.b) * (p - self.c))
    
    def obwod(self):
        return self.a + self.b + self.c
    
    def wysokosc(self):
        return 2 * self.pole() / self.a


class TrojkatProstokatny(Trojkat):
    def __init__(self, a, b):
        c = math.sqrt(a ** 2 + b ** 2)
        super().__init__(a, b, c)


class TrojkatRownoramienny(Trojkat):
    def __init__(self, a, b):
        super().__init__(a, b, b)
        
    def wysokosc(self):
        return math.sqrt(self.b ** 2 - (self.a / 2) ** 2)
    

class TrojkatRownoboczny(Trojkat):
    def __init__(self, a):
        super().__init__(a, a, a)
        
    def wysokosc(self):
        return math.sqrt(3) / 2 * self.a
```

```python
t1 = Trojkat(3, 4, 5)
t2 = TrojkatProstokatny(3, 4)
t3 = TrojkatRownoramienny(4, 5)
t4 = TrojkatRownoboczny(3)

print("Obwody:")
print(t1.obwod(), t2.obwod(), t3.obwod(), t4.obwod())

print("Pola:")
print(t1.pole(), t2.pole(), t3.pole(), t4.pole())

print("Wysokości:")
print(t1.wysokosc(), t2.wysokosc(), t3.wysokosc(), t4.wysokosc())

print("Czy większy od:")
print(t1.czy_wiekszy_od(t2))
print(t2.czy_wiekszy_od(t3))
print(t3.czy_wiekszy_od(t4))
print(t4.czy_wiekszy_od(t1))

class Kwadrat:
    def __init__(self, a):
        self.a = a
    
    def pole(self):
        return self.a**2
    
    def obwod(self):
        return 4*self.a

class Prostokąt:
    def __init__(self, a, b):
        self.a = a
        self.b = b
    
    def pole(self):
        return self.a*self.b
    
    def obwod(self):
        return 2*self.a + 2*self.b

s1 = Kwadrat(5)
r1 = Prostokąt(3, 4)

print(t1.czy_wiekszy_od(s1))
print(t2.czy_wiekszy_od(r1))
print(t3.czy_wiekszy_od(s1))
print(t4.czy_wiekszy_od(r1))
```

### Zad PD0503
Pojawia się klasyk... Proszę zaprojektować hierarchię klas statków kosmicznych świata [StarWars](http://www.sluisvan.net/)

    StarWarsVehicle
    ...Mysliwce
    ...Transportowce
    ...Stacjonarne

Każda klasa powinna mieć 
- docstring w tylu Numpy
- 5 atrybutów obowiązkowych: nazwa, prędkość $v$, uzbrojenie, całkowita moc uzbrojenia $cmu$, wytrzymałość $w$
- 2 dodatkowe atrybuty zależne od podklasy, w przykładzie jest cena i hipernapęd, ale mogą być dowolne
- get/set
- `__str__` i `__repr__`
- funkcja `_szansa` - wielkość obliczająca jaka jest szansa starcia z przeciwnikiem, obliczana w/g wzoru: $S = 0.2 v + 0.5 cmu + 0.3 w$
- `__gt__`, `__eq__`, `__ge__`, `__lt__`, `__le__` -> do tych porównań wykorzystamy funkcję `_szansa`
- funkcja `atakuj(self, other)` -> wykorzytująca te powyższe i zwracająca STR czy walka danego statku (`self`) z innym (`other`) jest wygrana czy przegrana


Przykład

    >>> xw = Mysliwce('X-wing', 
            v=8, 
            uzbrojenie='4 działka laserowe (sprzężone) o mocy 180, 2 wyrzutnie torped (6 sztuk) o mocy 270',
            cmu=4*180+2*270
            w=120,
            hipernaped=True,
            cena=150000
            )
    >>> tie = Mysliwce('TIE', 
            v=8, 
            uzbrojenie='1 podwójne działko laserowe o mocy 90',
            cmu=90
            w=60,
            hipernaped=False,
            cena="NA"
            )
    >>> xw > tie
    True
    >>> xw.atakuj(tie)
    "Wygrana!"
    >>> tie._szansa()
    64.6
    
    
Na koniec proszę przeprowadzić testy klas: 
* 3 konkretne, różne instancje na klasę potomną, prosze przejrzeć funkcje które państwo programują
* walka każdy z każdym, najlepiej w pętli

```python
class StarWarsVehicle:
    def __init__(self, name, v, uzbrojenie, cmu, w):
        self.name = name
        self.v = v
        self.uzbrojenie = uzbrojenie
        self.cmu = cmu
        self.w = w
        
    def _szansa(self):
        return 0.2*self.v + 0.5*self.cmu + 0.3*self.w
    
    def atakuj(self, other):
        if self._szansa() > other._szansa():
            return "Wygrana!"
        elif self._szansa() < other._szansa():
            return "Przegrana!"
        else:
            return "Remis!"
        
    def __str__(self):
        return self.name
    
    def __repr__(self):
        return self.__class__.__name__ + "('" + self.name + "')"
    
    def __lt__(self, other):
        return self._szansa() < other._szansa()
    
    def __le__(self, other):
        return self._szansa() <= other._szansa()
    
    def __eq__(self, other):
        return self._szansa() == other._szansa()
    
    def __ge__(self, other):
        return self._szansa() >= other._szansa()
    
    def __gt__(self, other):
        return self._szansa() > other._szansa()
    
    def get_name(self):
        return self._name
    
    def get_v(self):
        return self._v
    
    def get_uzbrojenie(self):
        return self._uzbrojenie
    
    def get_cmu(self):
        return self._cmu
    
    def get_w(self):
        return self._w
    
    def set_name(self, name):
        self._name = name
        
    def set_v(self, v):
        self._v = v
        
    def set_uzbrojenie(self, uzbrojenie):
        self._uzbrojenie = uzbrojenie
        
    def set_cmu(self, cmu):
        self._cmu = cmu
        
    def set_w(self, w):
        self._w = w


class Mysliwce(StarWarsVehicle):
    def __init__(self, name, v, uzbrojenie, cmu, w, hipernaped, cena):
        super().__init__(name, v, uzbrojenie, cmu, w)
        self.hipernaped = hipernaped
        self.cena = cena
    
    def __str__(self):
        return f' nazwa:{self.name}, v:{self.v}, uzbrojenie:{self.uzbrojenie},\
            cmu:{self.cmu}, w:{self.w}, hipernaped:{self.hipernaped}, cena:{self.cena}'
    
    def get_hipernaped(self):
        return self._hipernaped
    
    def get_cena(self):
        return self._cena
    
    def set_hipernaped(self, hipernaped):
        self._hipernaped = hipernaped
        
    def set_cena(self, cena):
        self._cena = cena


class Transportowce(StarWarsVehicle):
    def __init__(self, name, v, uzbrojenie, cmu, w, ladownosc, cena):
        super().__init__(name, v, uzbrojenie, cmu, w)
        self.ladownosc = ladownosc
        self.cena = cena
        
    def __str__(self):
        return f' nazwa:{self.name}, v:{self.v}, uzbrojenie:{self.uzbrojenie},\
            cmu:{self.cmu}, w:{self.w}, ladownosc:{self.ladownosc}, cena:{self.cena}'
    
    def get_hipernaped(self):
        return self._hipernaped
    
    def get_cena(self):
        return self._cena
    
    def set_ladownosc(self, ladownosc):
        self._ladownosc = ladownosc
        
    def set_cena(self, cena):
        self._cena = cena


class Stacjonarne(StarWarsVehicle):
    def __init__(self, name, v, uzbrojenie, cmu, w, opancerzenie, cena):
        super().__init__(name, v, uzbrojenie, cmu, w)
        self.opancerzenie = opancerzenie
        self.cena = cena
     
    def __str__(self):
        return f' nazwa:{self.name}, v:{self.v}, uzbrojenie:{self.uzbrojenie},\
            cmu:{self.cmu}, w:{self.w}, opancerzenie:{self.opancerzenie}, cena:{self.cena}'
    
    def get_hipernaped(self):
        return self._hipernaped
    
    def get_cena(self):
        return self._cena
    
    def set_opancerzenie(self, opancerzenie):
        self._opancerzenie = opancerzenie
        
    def set_cena(self, cena):
        self._cena = cena
        
    def _szansa(self):
        return 0.2*self.v + 0.5*self.cmu + 0.3*self.w + 0.2*self.opancerzenie

```

### Zad PD05X. Generator docstringu Numpy
Napisz funkcję generującą docstring w typie Numpy, bazując na definicji funkcji `__init__`. Funkcja powinna pobierać zestaw parametrów w postaci słownika

    {'Parameters': None, 'Returns': None, 'Raises': None, 
    'Attributes': None, 'Summary': None, 'Description': None, 'Methods': None}

i generować poprawny docstring (w odpowiedniej kolejności).

    >>> print(gen_doc_numpy({'Parameters': ['a', 'b'], 
                'Attributes': ['a', 'b'], 
                'Methods': ['set_cena'],
                'Summary': "Obiekt Gruszka opisuje nam wlasnosci gruszek."}))
    '''
    Obiekt Gruszka opisuje nam wlasnosci gruszek.

    Parameters
    ---------- 
    a: type
        opis parametru a
    b: type
        opis parametru b
    
    Attributes
    ----------
    a: tu przechowujemy informacje o parametrze a
    b: tu przechowujemy informacje o parametrze b
    
    Methods
    -------
    set_cena(x)
        OPIS

    '''
    

```python
def gen_doc_numpy(info_dict):
    """
    Generuje docstring w formacie Numpy na podstawie podanych informacji.

    Parameters
    ----------
    info_dict : dict
        Słownik zawierający informacje o funkcji.

    Returns
    -------
    str
        Docstring w formacie Numpy.

    Raises
    ------
    None

    Attributes
    ----------
    None

    Summary
    -------
    str
        Podsumowanie opisujące funkcję.

    Description
    -----------
    str
        Pełny opis funkcji.

    Methods
    -------
    str
        Opis metod związanych z funkcją.

    """
    params = info_dict.get('Parameters')
    attrs = info_dict.get('Attributes')
    methods = info_dict.get('Methods')
    summary = info_dict.get('Summary')
    desc = info_dict.get('Description')

    doc = f"'''{summary}\n\n"
    if params:
        doc += "Parameters\n----------\n"
        for param in params:
            doc += f"{param}: type\n    opis parametru {param}\n"
    if attrs:
        doc += "\nAttributes\n----------\n"
        for attr in attrs:
            doc += f"{attr}: tu przechowujemy informacje o parametrze {attr}\n"
    if methods:
        doc += "\nMethods\n-------\n"
        for method in methods:
            doc += f"{method}\n    OPIS\n"
    doc += "'''"

    return doc
```

```python
print(gen_doc_numpy({'Parameters': ['a', 'b'], 
            'Attributes': ['a', 'b'], 
            'Methods': ['set_cena'],
            'Summary': "Obiekt Gruszka opisuje nam wlasnosci gruszek."}))
```
